```python
marks = [45, 78, 90, 33, 60]
students_pass=0
students_fail=0
for i in marks:
    if i>=50:
        students_pass+=1
        print("Pass")
    else:
        students_fail+=1 
        print("Fail")
print("Total students passed:", students_pass)
print("Total students failed:", students_fail)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
for i in range(len(names)):
    names[i]=names[i].strip().lower()
print(names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    print(f"Length of message: {len(message)}")
    if len(message) > 10:
        print(f"Flag message: {message}")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for i in range(len(logs)):
    if logs[i]=="ERROR":
        print("Error at index:", i)
        error_count+=1
print(f"Total ERRORS: {error_count}")
    
```
