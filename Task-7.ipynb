```python
#1
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False  

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")


my_light = SmartLight("Bedroom Light")


my_light.turn_on()


my_light.display_status()
```

```python

#2
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

emp1 = Employee("Rahul", "EMP102", "IT")


emp1.display_id_card()
```

```python
#3
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

contact1 = Contact("Anita", "9876543210")


contact1.display_contact()
```

```python
#4
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

item1 = Product("Headphones", 2499)

item1.print_price_tag()
```

```python
#5
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


movie1 = Movie("Inception", 4.8)

movie1.display_details()
```

```python
#6
class DeliveryRecord:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

record1 = DeliveryRecord("Suman", "Hyderabad")


record1.print_delivery_details()
```
