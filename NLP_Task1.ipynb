```python
def preprocess_text(text):
    if not text:
        return ""

    # Convert to lowercase
    text = text.lower()

    # 2. Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Normalize repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Remove very short tokens (≤2), keep exceptions
    exceptions = {"no", "not"}
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    cleaned=" ".join(tokens)

    return cleaned,tokens

text="I have 2 dogs and this is good, soooo goooood!!!, WOW!!! This is GREAT!!! now please Visit http://example.com"
cleaned,tokens=preprocess_text(text)
print("Tokens:",tokens)
print("Sentence:",cleaned)
```

```python
text=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

clean=[]
for t in text:
  cleaned,tokens=preprocess_text(t)

  print("\nOriginal Text:",t)
  print("Cleaned Tokens:",tokens)
  print("Cleaned Sentence:",cleaned)
  clean.append((tokens,cleaned))
```

```python
def full_pipeline(text_list):
  all_words=[]
  cleaned_Sentences=[]

  for t in text_list:
    cleaned,tokens=preprocess_text(t)
    cleaned_Sentences.append(cleaned)
    all_words.extend(tokens)

  print(f"Tokens:{all_words}")
  print(f"Sentences:{cleaned_Sentences}")
```
