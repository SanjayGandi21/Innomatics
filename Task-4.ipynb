```python
#3
def result(marks):
    tot=0
    for mark in marks:
        tot+=mark
    average=tot/len(marks)
    if average>=50:
        return "Pass"
    else:        
        return "Fail"

marks1=[45, 55, 60, 70, 80]
result(marks1)
marks2=[30, 40, 45, 35, 25]
result(marks2)
```

```python
#4
def fare_calculator(distance, peak_hour):
    peak_hour = peak_hour.lower()
    base_fare = 50
    total = base_fare + (distance * 12)
    if peak_hour == "yes":
        total*=1.25
    return total

while True:
    distance = float(input("Enter distance traveled (in km): "))
    peak_hour = input("Is it peak hour? (yes/no): ")
    fare = fare_calculator(distance, peak_hour)
    print(f"Total fare: ₹{fare:.2f}")
    retry = input("Do you want to retry? (yes/no): ")
    if retry.lower() != "yes":
        break

```

```python
#5
def attendance_percentage(attendance):
    tot=len(attendance)
    count=0
    for i in attendance:
        if i=="P":
            count+=1
    percentage=(count/tot)*100
    print(f"Attendance percentage: {percentage:.2f}%")
    if percentage>=75:
        return 'Eligible'
    else:
        return 'Not Eligible'

attendance1=["P", "A", "P", "P", "A", "P", "P", "A", "P", "P"]
attendance_percentage(attendance1)
```

```python
#6
def password_strength(password):
    if len(password) < 8:
        return "Weak Password,Enter at least 8 characters"
    has_letter=False
    has_digit=False
    has_special=False
    for char in password:
        if char.isalpha():
            has_letter=True
        elif char.isdigit():
            has_digit=True
        elif not char.isalnum():
            has_special=True
    
    if has_letter and has_digit and has_special:
        return "Strong Password"
    elif has_letter and has_digit:
        return "Moderate Password,Add special characters for a stronger password"
    elif has_letter and has_special:
        return "Moderate Password,Add digits for a stronger password"
    elif has_digit and has_special:
        return "Moderate Password,Add letters for a stronger password"
    else:
        return "Weak Password,Include letters, digits, and special characters for a stronger password"
        

Password=input("Enter a password: ")
password_strength(Password)

        
```
