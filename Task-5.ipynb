```python
#1
def post_status(likes):
    if likes>=1000:
        return "Viral Post"
    else: return "Normal Engagement"

likes=int(input("Enter the number of likes: "))
print("Total likes:",likes)
print(f"Post Status: {post_status(likes)}")
```

```python
#2
def stock_management(stock):
    if stock<10:
        return "Low Stock Alert!"
    elif stock>=10:
        return "Stock Sufficient"

stock=int(input("Enter the current stock level: "))
print("Current stock level:",stock)
print(f"Stock Status: {stock_management(stock)}")
```

```python
#3
def rainfall_checker(rainfall,required_level):
    tot=0
    for i in rainfall:
        tot+=i
    avg=tot/len(rainfall)
    print(f"Average Rainfall: {avg}")
    if avg>=required_level:
        return "Adequate Rainfall"
    else: return "Inadequate Rainfall"

rainfall=[float(input("Enter rainfall for day {}: ".format(i+1))) for i in range(7)]
required_level=float(input("Enter the required rainfall level: "))
print(f"Rainfall Status: {rainfall_checker(rainfall,required_level)}")
```

```python
#4
def duplicate_checker(lst):
    seen=set()
    for item in lst:
        if item in seen:
            return "Yes"
        seen.add(item)
    return "No"

lst=["Sanjay","spidy","Hamza","Batman","Sanjay","Husky","Ojas Gambheera","Hamza"]
print(f"Duplicate Found: {duplicate_checker(lst)}")
```

```python
#5
def eligibility_checker(age):
    if age>=18:
        return "Eligible"
    else: return "Not Eligible"
age=int(input("Enter your age: "))
print("Patient age:",age)
print(f"Eligibility Status: {eligibility_checker(age)}")
```

```python
#6
def crop_filter(prices):
    premium=[]
    regular=[]
    for price in prices:
        if price>2000:
            premium.append(price)
        else:
            regular.append(price)
    return f"Premium Crops: {premium}, Regular Crops: {regular}"

prices=list(map(int,input("Enter crop prices separated by space: ").split()))
print(crop_filter(prices))

    
```

```python
#7
def application_error(error):
    if error==0:
        return "Healthy"
    elif error<=5:
        return "Minor Issues"
    elif error>5:
        return "Critical Issues"

error=int(input("Enter the number of errors: "))
print(f"Error Count: {error}")
print(f"System Status: {application_error(error)}")
```

```python
#8
def daily_limit(transactions):
    if transactions<0:
        return "Invalid Transaction Amount"
    elif transactions<=50000:
        return "Transaction Approved"
    else: return "Transaction Rejected"

transactions=(int(input("Enter the total transaction amount")))
print(f"Total Transaction Amount: {transactions}")
print(f"Transaction Status: {daily_limit(transactions)}")


```

```python
#9
def student_attendance(attendance):
    if attendance>=75:
        return "Eligible"
    else: return "Not Eligible"

attendance=int(input("Enter your attendance percentage: "))
print("Attendance Percentage:",attendance)
print(f"Exam Status: {student_attendance(attendance)}")
```

```python
def electricity_bill(units):
    bill=0
    if units<=100:
        bill=units*3
        print(f"Total bill: {bill}")
    elif units<=200:
        bill=units*5
        print(f"Total bill: {bill}")
    else:
        bill=units*7
        print(f"Total bill: {bill}")
    
    if bill<500:
        return "Low Usage"
    elif bill>500 and bill<=1500:
        return "Moderate Usage"
    elif bill>1500:
        return "High Usage"
    
    
units=int(input("Enter the number of electricity units consumed: "))
print("Total Units Consumed:",units)
print(f"Electricity Usage:{electricity_bill(units)}")
```
