```python
#1
sentence=input("Enter a sentence: ")
unique_words=set(sentence.split())
print(f"Unique words:{unique_words}")
print(f"Number of unique words:{len(unique_words)}")
```

```python
#2
employees = {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
highest=max(employees.values())
for key,value in employees.items():
    if value==highest:
        print(f"Highest Salary:{key}-{value}")

```

```python
#3
numbers = [45, 22, 89, 10, 66]
max=numbers[0]
min=numbers[0]
for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num
print(f"List:{numbers}")
print(f"Maximum number:{max}")
print(f"Minimum number:{min}")
```

```python
#4
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
    if i>1000:
        count+=1
print(f"Products above 1000:{count}")
```

```python
#5
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i=="P":
        count+=1
Percentage=(count/len(attendance))*100
print(f"Attendance percentage:{Percentage}%")
```

```python
#6
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers=set(phone_numbers)
print(f"Unique phone numbers:{unique_numbers}")
```

```python
#7
text ="Python"
dict={}
for char in text:
    if char in dict:
        dict[char]+=1
    else:
        dict[char]=1
print(f"Character frequency of {text}:{dict}")
```
