```python
#1
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}
high=max(employees.values())
top_performers=[name for name,score in employees.items() if score==high]
print(f"Top Performers Eligible for Bonus:{', '.join(top_performers)} (Score: {high})")

```

```python
#2
from collections import Counter
q="Buy mobile phone buy phone online"
q=q.lower()
counts=Counter(q.split())
ecom={word:count for word,count in counts.items() if count>1}
print(ecom)
```

```python
#3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
f=[]
for i in range(len(sensor_readings)):
    if sensor_readings[i]%2==0:
        h=(i,sensor_readings[i])
        f.append(h)
print(f"Valid Sensor Readings (Hour, Value):{f}")
  
```

```python
#4
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
g=0
y=0
for i in emails:
    if i.endswith("gmail.com"):
        g+=1
    elif i.endswith("yahoo.com"):
        y+=1
gmail=g/len(emails)*100
yahoo=y/len(emails)*100
print(f"gamil.com:{gmail}%")
print(f"yahoo.com:{yahoo}%")
```

```python
#5
sales = [1200, 1500, 900, 2200, 1400, 3000]
s=sum(sales)
avg=s/len(sales)
pe=avg*1.3
for i in range(len(sales)):
    if sales[i]>=pe:
        print(f"Day{i}: {sales[i]}")
```
