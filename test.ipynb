```python
import pandas as pd

orders_df = pd.read_csv("orders.csv")
users_df = pd.read_json("users.json")
restaurants_df = pd.read_csv("restaurants.csv")

print(orders_df.head())

orders_with_users = pd.merge(
    orders_df,
    users_df,
    on='user_id',
    how='left'
)

merged_df = pd.merge(
    orders_with_users,
    restaurants_df,
    on='restaurant_id',
    how='left'
)

print(merged_df.columns)
print(merged_df.head())

merged_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
total_orders = len(merged_df)
gold_orders = merged_df[merged_df['membership'] == 'Gold']
count_gold = len(gold_orders)

percentage = (count_gold / total_orders) * 100
print(round(percentage))
```

```python
grouped_data = merged_df.groupby(['membership', 'cuisine'])
revenue_stats = grouped_data['total_amount'].sum()

print(revenue_stats.sort_values(ascending=False))
```

```python
merged_df['order_date'] = pd.to_datetime(merged_df['order_date'], dayfirst=True)
merged_df['quarter'] = merged_df['order_date'].dt.quarter

quarter_revenue = merged_df.groupby('quarter')['total_amount'].sum()
print(quarter_revenue.sort_values(ascending=False))
```

```python
gold_members = merged_df[merged_df['membership'] == 'Gold']
print(len(gold_members))
```

```python
hyderabad_data = merged_df[merged_df['city'] == 'Hyderabad']
total = hyderabad_data['total_amount'].sum()

print(round(total))
```

```python
distinct_users = merged_df['user_id'].nunique()
print(distinct_users)
```

```python
gold_data = merged_df[merged_df['membership'] == 'Gold']
average_val = gold_data['total_amount'].mean()

print(round(average_val, 2))
```

```python
high_rating = merged_df[merged_df['rating'] >= 4.5]
print(len(high_rating))
```
